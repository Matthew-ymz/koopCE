```python
import pysr

import sympy
import numpy as np
from matplotlib import pyplot as plt
from pysr import PySRRegressor
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
X = 2 * np.random.randn(100, 5)
y = 2.5382 * np.cos(X[:, 3]) + X[:, 0] ** 2 - 2

default_pysr_params = dict(
    populations=40,
    model_selection="best",
)

model = PySRRegressor(
    niterations=30,
    binary_operators=["+", "*"],
    unary_operators=["cos", "exp", "sin", 'square', 'abs', 'log2', 'tanh', 'cosh', 'sinh', 'sqrt'],
    **default_pysr_params,
)

model.fit(X, y)


model.sympy()
```

```python
np.random.seed(0)
X = 2 * np.random.randn(100, 5)
y = 2.5382 * np.cos(X[:, 3]) + X[:, 0] ** 2 - 2

default_pysr_params = dict(
    populations=40,
    model_selection="best",
)

model = PySRRegressor(
    niterations=30,
    binary_operators=["+", "*"],
    unary_operators=["cos", "exp", "sin", "pow(x) = x^2"],
    extra_sympy_mappings={"pow": lambda x: x**2},
    **default_pysr_params,
)

model.fit(X, y)
model.sympy()
```
